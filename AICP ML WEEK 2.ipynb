```python
#  / / Question No. 2 \ \
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
tips_df = pd.read_csv("E:\MNS-UET BS IT\AICP ML INTERNSHIP\ML Internship Task 2/tips.csv")

# Plotting tips vs. total bill paid
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips_df)
plt.title('Tips vs Total Bill Paid')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.grid(True)
plt.show()

# Plotting tips vs. number of people at a table
plt.figure(figsize=(8, 6))
sns.barplot(x='size', y='tip', data=tips_df)
plt.title('Tips vs Number of People at a Table')
plt.xlabel('Number of People')
plt.ylabel('Tip ($)')
plt.grid(True)
plt.show()

# Plotting tips vs. the day of the week
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='tip', data=tips_df)
plt.title('Tips vs Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Tip ($)')
plt.grid(True)
plt.show()
```

```python
#  / / Question No. 3 \ \
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of seaborn
sns.set(style="whitegrid")

# Create subplots to display the visualizations
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Visualize tips vs. total bill paid by gender
sns.scatterplot(data=tips_df, x='total_bill', y='tip', hue='sex', ax=axes[0])
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Tip')
axes[0].set_title('Tips vs. Total Bill by Gender')

# Visualize tips vs. number of people at a table by gender
sns.barplot(data=tips_df, x='size', y='tip', hue='sex', ax=axes[1])
axes[1].set_xlabel('Number of People at a Table')
axes[1].set_ylabel('Tip')
axes[1].set_title('Tips vs. Number of People at a Table by Gender')

# Visualize distribution of tips by gender
sns.boxplot(data=tips_df, x='sex', y='tip', ax=axes[2])
axes[2].set_xlabel('Gender')
axes[2].set_ylabel('Tip')
axes[2].set_title('Distribution of Tips by Gender')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
#  / / Question No. 5 \ \
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate total tips received on each day
total_tips_per_day = tips_df.groupby('day')['tip'].sum().reset_index()

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a bar plot to visualize total tips received on each day
plt.figure(figsize=(8, 6))
sns.barplot(data=total_tips_per_day, x='day', y='tip', hue='day', palette='viridis', legend=False)
plt.xlabel('Day of the Week')
plt.ylabel('Total Tips')
plt.title('Total Tips Given to Waiters on Each Day')
plt.show()
```

```python
#  / / Question No. 6 \ \
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate total tips received by each gender
total_tips_per_gender = tips_df.groupby('sex')['tip'].count().reset_index()

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a bar plot to visualize total tips given by each gender
plt.figure(figsize=(6, 6))
sns.barplot(data=total_tips_per_gender, x='tip', y='sex', hue='sex', palette='muted', legend=False)
plt.xlabel('Number of Tips')
plt.ylabel('Gender')
plt.title('Number of Tips Given to Waiters by Gender')
plt.yticks(ticks=[0, 1], labels=['Male', 'Female'])
plt.show()
```

```python
#  / / Question No. 8 \ \
import matplotlib.pyplot as plt

# Calculate average tip amount for smokers and non-smokers
avg_tip_smoker = tips_df[tips_df['smoker'] == 'Yes']['tip'].mean()
avg_tip_non_smoker = tips_df[tips_df['smoker'] == 'No']['tip'].mean()

# Set the size of the figure
plt.figure(figsize=(6, 4))

# Create a bar plot to visualize average tip amount for smokers and non-smokers
plt.bar(['Smoker', 'Non-Smoker'], [avg_tip_smoker, avg_tip_non_smoker], color=['orange', 'green'])
plt.xlabel('Smoking Status')
plt.ylabel('Average Tip Amount')
plt.title('Average Tip Amount: Smoker vs. Non-Smoker')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#  / / Question No. 9 \ \
import matplotlib.pyplot as plt

# Calculate total tips received during lunch and dinner
total_tips_lunch = tips_df[tips_df['time'] == 'Lunch']['tip'].sum()
total_tips_dinner = tips_df[tips_df['time'] == 'Dinner']['tip'].sum()

# Set the size of the figure
plt.figure(figsize=(6, 4))

# Create a bar plot to visualize total tips received during lunch and dinner
plt.bar(['Lunch', 'Dinner'], [total_tips_lunch, total_tips_dinner], color=['skyblue', 'salmon'])
plt.xlabel('Time of Meal')
plt.ylabel('Total Tips')
plt.title('Total Tips Received: Lunch vs. Dinner')
plt.tight_layout()

# Show the plot
plt.show()
```
